```python
FIGSIZE_WIDTH = 12
FIGSIZE_HEIGHT = 6
DPI = 150

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

df = pd.read_excel('Mappe1.xlsx')
mean = df.mean(axis=0)
mode = df.mode()
median = df.median(axis=0)

for col in df.drop('Teilnehmer', axis=1).columns:
    fig, ax = plt.subplots(figsize=(FIGSIZE_WIDTH, FIGSIZE_HEIGHT), dpi=DPI)
    get_dens = gaussian_kde(df[col]).evaluate
    df[col].plot(kind='kde', ax=ax)

    ax.vlines(mean[col], 0, get_dens(mean[col]), label='Mittelwert', color='red', linestyles='dashed')
    ax.vlines(median[col], 0, get_dens(median[col]), label='Median', color='orange', linestyle='dashdot')
    ax.vlines(mode[col][0], 0, get_dens(mode[col][0]), label='Modus', color='green', linestyles='dotted')

    ax.legend()
    ax.set_ylim(0)
    ax.set_title(col)
    plt.show()
```
